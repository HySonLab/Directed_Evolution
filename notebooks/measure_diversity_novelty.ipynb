```python
from polyleven import levenshtein
import numpy as np
import pandas as pd
from de.predictors.tranception.utils.scoring_utils import get_mutated_sequence
```

```python
filepath = "/home/thanhtvt1/workspace/Directed_Evolution/exps/results/results_avGVF_steps60_pop128_pros4_seed0.csv"
df = pd.read_csv(filepath)
df
```

```python
wt_seq = df["WT"].to_list()[0]
mutants = df["mutants"].to_list()
seqs = []
for mut in mutants:
    seqs.append(get_mutated_sequence(wt_seq, mut))
print(len(seqs))
```

```python
from itertools import combinations


def edit_distance(seq1, seq2):
    return levenshtein(seq1, seq2)


def measure_diversity(seqs):
    dists = []
    for pair in combinations(seqs, 2):
        dists.append(edit_distance(*pair))
    return np.mean(dists)


def measure_novelty(seqs, wt):
    dists = []
    for seq in seqs:
        dists.append(edit_distance(seq, wt))
    return np.mean(dists)
```

```python
diversity = measure_diversity(seqs)
novelty = measure_novelty(seqs, wt_seq)

print("Diversity:", diversity)
print("Novelty  :", novelty)
```
